```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python

X_train
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ('tr1',SimpleImputer(),['fever']), #filling missing values
    ('tr2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']), #Ordinal data(Categorical data) --> OrdinalEncoder
    ('tr3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city']) #Nominal data(Categorical data) --> OneHotEncoder
],remainder='passthrough') #Age would be passthrough
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
