# 4. Distplot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5. Heatmap (Categorical-Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
titanic.groupby('Pclass').mean('Survived')['Survived']*100 #Percentages who survived
```

```python
(titanic.groupby('Pclass').mean('Survived')['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex').mean('Survived')['Survived']*100).plot(kind='bar') # Females survived more
```

```python
(titanic.groupby('Embarked').mean('Survived')['Survived']*100).plot(kind='bar') #Why chibhog people survived more? More females or first-class people?
```

# 6. Clustermap (Categorical-Categorical)
