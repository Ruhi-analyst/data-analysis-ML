```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

# To predict whether the passenger survived or not

```python
df.head()
```

# 1. Dropping Columns

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)     #df.drop(,inplace=True)
```

# 2. Train Test Split

```python
X_train,X_test,y_train,y_test=train_test_split(
    df.drop(columns=['Survived']),
    df['Survived'],
    test_size=0.2,
    random_state=42
)
```

```python
X_train.head()
```

```python
y_train.head()
```

# 3. Imputation Transformer

```python
tr1=ColumnTransformer([
    ('Imp_Age',SimpleImputer(),[2]),
    ('Imp_Emb',SimpleImputer(strategy='most_frequent'),[6])     
    ],remainder='passthrough')
```

# 4. OHE Transformer

```python
tr2=ColumnTransformer([
    ('ohe_sex_emb',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
    ],remainder='passthrough')
```

# 5. Scaling

```python
tr3= ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
     
     ])
```

# 6. Feature Selection

```python
tr4=SelectKBest(score_func=chi2,k=8)
```

# 7. Train the model

```python
trf5 = DecisionTreeClassifier()
```

# Create pipleline

```python
pipe=make_pipeline(tr1,tr2,tr3,tr4,trf5)
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# train
pipe.fit(X_train,y_train)
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python

y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

# GridSearch -Fine Tuning

```python
# gridsearchcv
params = {
    'decisiontreeclassifier__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```
